```python
# from https://medium.com/@yulongtsai/https-medium-com-yulongtsai-titanic-top3-8e64741cc11f
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set(font_scale=1.56)
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
df_data = df_train.append(df_test)
```

```python
sns.countplot(df_data["Sex"], hue=df_data["Survived"])
display(df_data[["Sex", "Survived"]].groupby(["Sex"], as_index=False).mean().round(3))
```

```python
sns.countplot(df_data["Age"], hue=df_data["Survived"])
display(df_data[["Age", "Survived"]].groupby(["Age"], as_index=False).mean().round(3))
```
